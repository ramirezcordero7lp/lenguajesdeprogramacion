# Caso práctico: Analista de inversiones

Imagina que trabajas como **analista financiero** en una empresa de inversión.

Tu equipo debe decidir en cuál proyecto invertir.

La empresa solo puede elegir **un proyecto**.

Tu tarea será usar **Python para analizar la rentabilidad de cada proyecto** utilizando el **Valor Presente Neto (VPN)**.

## ¿Qué es el Valor Presente Neto (VPN)?

El **Valor Presente Neto (VPN)** permite determinar si un proyecto de inversión es rentable.

- Si **VPN > 0** → el proyecto genera valor.
- Si **VPN < 0** → el proyecto no es rentable.

Fórmula:

$$VPN = \\sum \
rac{Flujo_t}{(1+r)^t} - Inversión\ Inicial$$

## Proyectos de inversión

Usaremos **diccionarios en Python** para representar cada proyecto.

```python
proyecto_Carlos= {
    'nombre': 'Proyecto Carlos Ltda',
    'inversion': 8000,
    'flujos': [2500, 3000, 3500, 4000],
    'tasa': 0.10
}

proyecto_Luisa = {
    'nombre': 'Proyecto Luisa SAS',
    'inversion': 10000,
    'flujos': [3000, 3500, 4200, 4500],
    'tasa': 0.10
}

proyecto_Liz = {
    'nombre': 'Proyecto Liz SAS',
    'inversion': 7000,
    'flujos': [1800, 2600, 3200, 3800],
    'tasa': 0.10
}
```

## Calcular el VPN usando `for`

## Ejercicio

Calcule el VPN del **Proyecto B** y **Proyecto C**.

```python
vpn = -proyecto_Carlos['inversion']

for t, flujo in enumerate(proyecto_Carlos['flujos']):
    vpn += flujo / (1 + proyecto_Carlos['tasa'])**(t+1)

print('VPN del proyecto A:', round(vpn, 3))
```

```python
if vpn > 25680:
  print('Muy bien tu proyecto es rentable felicitaciones.')
else:
    print('Ojo que tu proyecto no es rentable, por lo tanto te invito a buscar otra alternativa')
```

## Analizar todos los proyectos

```python
proyectos = [proyecto_Carlos, proyecto_Luisa, proyecto_Liz]

for proyecto in proyectos:
    vpn = -proyecto['inversion']

    for t, flujo in enumerate(proyecto['flujos']):
        vpn += flujo / (1 + proyecto['tasa'])**(t+1)

    print(proyecto['nombre'], 'VPN:', round(vpn, 2))
```

## Simulación de tasas con `while`

```python
tasa = 0.05

while tasa <= 0.20:
    vpn = -proyecto_Carlos['inversion']

    for t, flujo in enumerate(proyecto_Carlos['flujos']):
        vpn += flujo / (1 + tasa)**(t+1)

    print('Tasa:', tasa, 'VPN:', round(vpn, 2))

    tasa += 0.05
```
